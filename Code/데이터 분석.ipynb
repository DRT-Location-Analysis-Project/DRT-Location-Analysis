#### 자카드 유사도

```python
from itertools import combinations

# 자카드 유사도 함수 정의
def jaccard_similarity(set1, set2):
    intersection = len(set1.intersection(set2)) # set1과 set2의 교집합 크기 계산
    union = len(set1.union(set2)) # set1과 set2의 합집합 크기 계산
    return intersection / union # 교집합 크기를 합집합 크기로 나누어 자카드 유사도 반환

# 모든 자카드 결과를 저장할 리스트
all_jaccard_results = []

# 서울의 각 구 이름 리스트
gu = [
    "강남", "강동", "강북", "강서", "관악", "광진", "구로", "금천",
    "노원", "도봉", "동대문", "동작", "마포", "서대문", "서초", "성동",
    "성북", "송파", "양천", "영등포", "용산", "은평", "종로", "중랑"
]

# 각 구에 대해 반복
for i in gu:
    bus.loc[bus['노선번호'].str.contains(i), '지역번호'] = i  # '노선번호'에 현재 구 이름이 포함된 행의 '지역번호' 컬럼에 구 이름 할당
    filtered_bus = bus[bus['지역번호'] == i] # 현재 구 이름과 일치하는 '지역번호'를 가진 데이터 필터링
    # 필터링된 데이터프레임이 비어 있으면 다음 구로 넘어감
    if filtered_bus.empty:
        continue

    # '노선번호'별로 그룹화하여 '표준버스정류장ID'를 집합(set)으로 변환
    grouped = filtered_bus.groupby('노선번호')['표준버스정류장ID'].apply(set).reset_index()

    # 모든 조합에 대해 자카드 유사도 계산
    jaccard_results = []

    # 두 노선번호씩 조합하여 자카드 유사도 계산
    for (id1, set1), (id2, set2) in combinations(grouped.values, 2):
        similarity = jaccard_similarity(set1, set2) # 자카드 유사도 함수 이용해 계산
        jaccard_results.append([i, id1, id2, similarity]) # 결과를 리스트에 추가

    # 결과를 전체 결과 리스트에 추가
    all_jaccard_results.extend(jaccard_results)

# 최종 결과를 데이터프레임에 저장
final_jaccard_df = pd.DataFrame(all_jaccard_results, columns=['구', '노선번호1', '노선번호2', '자카드유사도'])

# 결과 출력
final_jaccard_df
```

```python
# '자카드 알고리즘 결과'가 0.25(25% 이하인 것 드롭)
final_jaccard_df = final_jaccard_df[final_jaccard_df['자카드유사도'] > 0.25]

# '자카드 알고리즘 결과' 열을 기준으로 내림차순 정렬
final_jaccard_df = final_jaccard_df.sort_values(by='자카드유사도', ascending=False)

# 필요없는 정보 삭제
final_jaccard_df = final_jaccard_df.reset_index(drop=True)
del final_jaccard_df['구']
```

#### 코사인 유사도

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

df = bus

# '표준버스정류장ID'를 사용하여 벡터화
def create_route_vector(route_df, all_stops):
    vector = [0] * len(all_stops)
    for stop in route_df['표준버스정류장ID']:
        index = all_stops.index(stop)
        vector[index] = 1
    return vector

# '표준버스정류장ID' 리스트로 정렬
all_stops = sorted(df['표준버스정류장ID'].unique())

# 각 노선번호에 대해 '표준버스정류장ID' 벡터화 된 것 적용
route_vectors = {}
route_ids = df['노선번호'].unique()
for route_id in route_ids:
    route_df = df[df['노선번호'] == route_id]
    route_vectors[route_id] = create_route_vector(route_df, all_stops)

# 각 노선번호의 벡터화 값을 이용해 코사인 유사도 계산
route_vectors_list = list(route_vectors.values())
cosine_matrix = cosine_similarity(route_vectors_list)

# 유사도 결과를 저장
cosine_results = []
for i in range(len(route_ids)):
    for j in range(i + 1, len(route_ids)):
        cosine_results.append({
            '노선번호1': route_ids[i],
            '노선번호2': route_ids[j],
            '코사인유사도': cosine_matrix[i, j]
        })

# 결과를 데이터프레임으로 변환
df_cosine = pd.DataFrame(cosine_results)

# 코사인 유사도를 기준으로 내림차순 정렬하고, 노선번호1을 오름차순으로 정렬
df_cosine = df_cosine.sort_values(by=['코사인유사도', '노선번호1'], ascending=[False, True]).reset_index(drop=True)

df_cosine.head(30)
```

```python
df_cosine.to_csv('df_cosine.csv')
```

```python
df = df[(df['사용년월'].apply(lambda x : x[:4]) == '2023') | (df['사용년월'].apply(lambda x : x[:4]) == '2024')]
```

```python
def yearly_person(bus):
    # 특정 노선번호 필터링
    df_bus = df[df['노선번호'] == bus]

    # 시간대별 승차총승객수 열만 선택 - 이용량 확인 가능
    승차총승객수_columns = [col for col in df_bus.columns if '승차총승객수' in col]

    # 년도 컬럼 추가
    df_bus['년도'] = df_bus['사용년월'].apply(lambda x: x[:4])

    # 시간대별 평균 이용량 계산
    df_avg = df_bus.groupby('년도')[승차총승객수_columns].mean().transpose()

    # 시각화
    plt.figure(figsize=(12, 8))

    for year in ['2023', '2024']:
        if year in df_avg.columns:
            hours = range(24)
            values = df_avg[year].values
            plt.plot(hours, values, marker='o', label=f'{year}')

    plt.xlabel('Hour')
    plt.ylabel('Average Persons')
    plt.title('Yearly Hourly Average Persons')
    plt.legend(title='Year')
    plt.grid(True)
    plt.xticks(range(24))
    plt.show()
```

#### 막대그래프 (평/토/일(공) 평균),  선그래프 (시간대별 평균 재차인원) 함수화

```python
import matplotlib.pyplot as plt

def average_three_range(data, route_name):
    route_data = data[data['노선명'] == route_name]

    # 평일, 토요일, 일요일(공휴일) 별 평균 재차인원 계산
    weekday_avg = route_data[route_data['weekday'] == True].iloc[:, 6:-3].mean()
    saturday_avg = route_data[route_data['saturday'] == True].iloc[:, 6:-3].mean()
    sun_holiday_avg = route_data[route_data['sun_holiday'] == True].iloc[:, 6:-3].mean()

    # 시간대별 평균 재차인원 계산
    time_columns = [f'{i}시' for i in range(24)]
    time = [i for i in range(24)]
    time_avg_weekday = route_data[route_data['weekday'] == True][time_columns].mean()
    time_avg_saturday = route_data[route_data['saturday'] == True][time_columns].mean()
    time_avg_sun_holiday = route_data[route_data['sun_holiday'] == True][time_columns].mean()

    # 그래프 1:  평일, 토요일, 일요일(공휴일) 일평균 재차인원
    plt.figure(figsize=(4,5))
    plt.bar(['Weekday', 'Saturday', 'Sunday/Holiday'],
            [weekday_avg.mean(), saturday_avg.mean(), sun_holiday_avg.mean()],
            color=['blue', 'green', 'red'])
    plt.xlabel('Day Type')
    plt.ylabel('Average Boarding')
    plt.show()

    print()

    # 그래프 2 : 시간대별 평균 재차인원
    plt.figure(figsize=(8, 5))
    plt.plot(time, time_avg_weekday, marker='o', label='Weekday', color = 'blue')
    plt.plot(time, time_avg_saturday, marker='o', label='Saturday',color = 'green')
    plt.plot(time, time_avg_sun_holiday, marker='o', label='Sunday/Holiday', color = 'red')
    plt.xlabel('Time of Day')
    plt.ylabel('Average Boarding')
    plt.xticks(range(24))
    plt.legend()
    plt.grid(True)
    plt.show()

    # 데이터프레임 생성 및 비율 계산
    df = pd.DataFrame({
        'Time': time_columns,
        'Weekday_Avg': time_avg_weekday,
        'Saturday_Avg': time_avg_saturday,
        'Sunday_Holiday_Avg': time_avg_sun_holiday
    })

    # 반올림
    df['평일_평균재차인원'] = round(df['Weekday_Avg'],2)
    df['토_평균재차인원'] = round(df['Saturday_Avg'],2)
    df['일(공)_평균재차인원'] = round(df['Sunday_Holiday_Avg'],2)

    # 데이터프레임 저장
    df.to_csv(f"{route_name}_평토일.csv")

    # 데이터가 없는 새벽 시간은 제거
    output = df.iloc[5:,4:]
    return output
```

#### 평일, 토, 일(공휴일) 이용량 확인 - api가 최신 데이터라 기간이 2023-07 ~ 2024-06 이므로 평균내서 2023년에 적용

```python
import pandas as pd
df01_ratio = pd.read_csv("강서01_평토일.csv")
df02_ratio = pd.read_csv("강서02_평토일.csv")
```

```python
from datetime import datetime, timedelta

# 공휴일 리스트
holidays = [
    '2023-01-01', '2023-01-21', '2023-01-22', '2023-01-23', '2023-01-24',
    '2023-03-01', '2023-05-27', '2023-05-29', '2023-06-06', '2023-08-15',
    '2023-09-28', '2023-09-29', '2023-09-30', '2023-10-02', '2023-10-03',
    '2023-10-09', '2023-12-25'
]

holidays = [datetime.strptime(date, '%Y-%m-%d') for date in holidays]

# 각 달의 평일, 토요일, 일요일(공휴일) 계산
def count_days(year):
    month_days = {i: {'weekdays': 0, 'saturdays': 0, 'sundays_holidays': 0} for i in range(1, 13)}
    for month in range(1, 13):
        first_day = datetime(year, month, 1)
        if month < 12 :
            last_day = datetime(year, month + 1, 1) - timedelta(days=1)
        else :
            datetime(year, 12, 31)
        current_day = first_day

        # 월별 평일. 토요일, 일요일(공휴일) 수
        while current_day <= last_day:
            if current_day.weekday() < 5 and current_day not in holidays:
                month_days[month]['weekdays'] += 1
            elif current_day.weekday() == 5:
                month_days[month]['saturdays'] += 1
            elif current_day.weekday() == 6 or current_day in holidays:
                month_days[month]['sundays_holidays'] += 1
            current_day += timedelta(days=1)

    return month_days

month_days = count_days(2023)

# 각 달의 일수 합산
total_weekdays = sum(days['weekdays'] for days in month_days.values())
total_saturdays = sum(days['saturdays'] for days in month_days.values())
total_sundays_holidays = sum(days['sundays_holidays'] for days in month_days.values())
```

```python
# 각 요일의 총 이용량 계산
df01_ratio['평일_총이용량'] = df01_ratio['평일_평균재차인원'] * total_weekdays
df01_ratio['토_총이용량'] = df01_ratio['토_평균재차인원'] * total_saturdays
df01_ratio['일(공)_총이용량'] = df01_ratio['일(공)_평균재차인원'] * total_sundays_holidays

# 전체 이용량 계산
df01_ratio['전체_이용량'] = df01_ratio['평일_총이용량'] + df01_ratio['토_총이용량'] + df01_ratio['일(공)_총이용량']

# 전체 대비 각 항목들의 비율 계산
df01_ratio['평일_비율'] = round(df01_ratio['평일_총이용량'] / df01_ratio['전체_이용량'],2)
df01_ratio['토_비율'] = round(df01_ratio['토_총이용량'] / df01_ratio['전체_이용량'],2)
df01_ratio['일(공)_비율'] = 1- (df01_ratio['평일_비율'] + df01_ratio['토_비율'])

# 결과 출력
df01 = df01_ratio[['평일_평균재차인원', '토_평균재차인원', '일(공)_평균재차인원', '평일_비율', '토_비율', '일(공)_비율']]
```

```python
# 각 요일의 총 이용량 계산
df02_ratio['평일_총이용량'] = df02_ratio['평일_평균재차인원'] * total_weekdays
df02_ratio['토_총이용량'] = df02_ratio['토_평균재차인원'] * total_saturdays
df02_ratio['일(공)_총이용량'] = df02_ratio['일(공)_평균재차인원'] * total_sundays_holidays

# 전체 이용량 계산
df02_ratio['전체_이용량'] = df02_ratio['평일_총이용량'] + df02_ratio['토_총이용량'] + df02_ratio['일(공)_총이용량']

# 전체 대비 각 항목들의 비율 계산
df02_ratio['평일_비율'] = round(df02_ratio['평일_총이용량'] / df02_ratio['전체_이용량'],2)
df02_ratio['토_비율'] = round(df02_ratio['토_총이용량'] / df02_ratio['전체_이용량'],2)
df02_ratio['일(공)_비율'] = 1- (df02_ratio['평일_비율'] + df02_ratio['토_비율'])

# 결과 출력
df02 = df02_ratio[['평일_평균재차인원', '토_평균재차인원', '일(공)_평균재차인원', '평일_비율', '토_비율', '일(공)_비율']]
```

```python
import pandas as pd
from datetime import datetime, timedelta

# 공휴일 리스트
holidays = [
    '2023-01-01', '2023-01-21', '2023-01-22', '2023-01-23', '2023-01-24',
    '2023-03-01', '2023-05-27', '2023-05-29', '2023-06-06', '2023-08-15',
    '2023-09-28', '2023-09-29', '2023-09-30', '2023-10-02', '2023-10-03',
    '2023-10-09', '2023-12-25'
]

holidays = [datetime.strptime(date, '%Y-%m-%d') for date in holidays]

# 각 달의 평일, 토요일, 일요일(공휴일) 계산
def count_days(year):
    month_days = {i: {'weekdays': 0, 'saturdays': 0, 'sundays_holidays': 0} for i in range(1, 13)}
    for month in range(1, 13):
        first_day = datetime(year, month, 1)
        if month < 12 :
            last_day = datetime(year, month + 1, 1) - timedelta(days=1)
        else :
            datetime(year, 12, 31)
        current_day = first_day

        # 월별 평일. 토요일, 일요일(공휴일) 수
        while current_day <= last_day:
            if current_day.weekday() < 5 and current_day not in holidays:
                month_days[month]['weekdays'] += 1
            elif current_day.weekday() == 5:
                month_days[month]['saturdays'] += 1
            elif current_day.weekday() == 6 or current_day in holidays:
                month_days[month]['sundays_holidays'] += 1
            current_day += timedelta(days=1)
    return month_days

month_days = count_days(2023)
```

```python
# '사용년월'을 datetime 형식으로 변환
df['사용년월'] = pd.to_datetime(df['사용년월'], format='%Y%m')

# 각 달의 평일, 토요일, 일요일(공휴일) 일수
df['평일일수'] = df['사용년월'].apply(lambda x: month_days[x.month]['weekdays'])
df['토요일일수'] = df['사용년월'].apply(lambda x: month_days[x.month]['saturdays'])
df['일요일공휴일일수'] = df['사용년월'].apply(lambda x: month_days[x.month]['sundays_holidays'])

# 각 시간대별로 '0평일_1토_2일(공)' 컬럼의 값에 따라 승하차 데이터를 일수로 나눔
time_columns = [col for col in df.columns if '_승차' in col or '_하차' in col]
for col in time_columns:
    df[col] = df.apply(lambda row: round(row[col] / row['평일일수'],2)
    if row['0평일_1토_2일(공)'] == 0
    else (round(row[col] / row['토요일일수'],2)
        if row['0평일_1토_2일(공)'] == 1
          else round(row[col] / row['일요일공휴일일수'],2)), axis=1)
```

```python
df['사용년월'] = df['사용년월'].apply(lambda x : x[:-3])
```

```python
# 1대당 평균 이용량 확인 함수 구현
def calculate_avg_riders(df, df2):
    # 23시 이후 강서01의 차량은 절반으로 줄고, 강서02는 1대만 다닌다고 가정
    def night_buses(row, line_name):
        if line_name == '강서01':
            return row['시간당평균배차수(평일)'] / 2
        elif line_name == '강서02':
            return 1
        else:
            return row['시간당평균배차수(평일)']  # 기본값으로 평일 배차수 사용

    # 시간대별 배차수 매핑
    avg_bus_count = df2.set_index('노선명').to_dict(orient='index')

    # 데이터프레임에 시간대별 매핑되는 배차수로 나눠서 1대당 평균 이용량 확인
    updated_df = df.copy()
    for idx, row in updated_df.iterrows():
        line_name = row['노선번호']
        day_type = row['0평일_1토_2일(공)']
        if line_name in avg_bus_count:
            avg_buses = avg_bus_count[line_name]
            for hour in range(6, 24):
                hour_str = f'{hour:02d}시'
                if hour == 23:
                    avg_bus = night_buses(avg_buses, line_name)
                else:
                    if day_type == 0:  # 평일
                        avg_bus = avg_buses['시간당평균배차수(평일)']
                    elif day_type == 1:  # 토요일
                        avg_bus = avg_buses['시간당평균배차수(토)']
                    elif day_type == 2:  # 일요일 및 공휴일
                        avg_bus = avg_buses['시간당평균배차수(일(공))']

                if not np.isnan(row[f'{hour_str}_승차']):
                    updated_df.at[idx, f'{hour_str}_승차'] = round(row[f'{hour_str}_승차'] / avg_bus,2)
                if not np.isnan(row[f'{hour_str}_하차']):
                    updated_df.at[idx, f'{hour_str}_하차'] = round(row[f'{hour_str}_하차'] / avg_bus,2)

    return updated_df
```

## 강서01, 02 노선 시각화

```python
import pandas as pd
import folium
from folium import plugins


# 노선번호에 따라 색상 설정
colors = {'강서01': 'red', '강서02': 'purple'}
df['color'] = df['노선번호'].map(colors)

# Sort data within each group by 정류장순서
df = df.sort_values(by=['노선번호', '정류장순서'])

# folium 지도 생성 (중심 좌표는 데이터를 기준으로 설정)
m = folium.Map(location=[df['Y좌표'].mean(), df['X좌표'].mean()], zoom_start=15, width='100%', height='100%')

# CircleMarker를 사용하여 데이터 포인트 추가
for idx, row in df.iterrows():
    folium.CircleMarker(
        location=[row['Y좌표'], row['X좌표']],
        radius= 4,
        color=row['color'],
        fill=True,
        fill_color=row['color'],
        fill_opacity=0.5
    ).add_to(m)

# Connect points in each group based on 정류장순서
for route, group in df.groupby('노선번호'):
    points = group[['Y좌표', 'X좌표']].values.tolist()
    points.append(points[0])  # Connect last point back to the first point
    folium.PolyLine(
        locations=points,
        color=colors[route],
        weight=5,
        opacity=0.5,
        popup=f"Route: {route}"
    ).add_to(m)

# 지도 표시
m.save("kangseo_station.html")
m
```

## 승하차량 시각화

```python
import folium

def visualize_boarding_data(time, board_type, day_type):

    # 노선번호에 따라 색상 설정
    colors = {'강서01': 'red', '강서02': 'purple'}

    # 시간과 승하차 정보에 따라 컬럼 이름 생성
    column_name = f"{time}_{board_type}"

    # 주어진 요일 필터링
    filtered_data = df[df['0평일_1토_2일(공)'] == day_type]

    # 노선번호에 따라 색상 설정
    filtered_data['color'] = filtered_data['노선번호'].map(colors)

    # Sort data within each group by 정류장순서
    filtered_data = filtered_data.sort_values(by=['노선번호', '정류장순서'])

    # folium 지도 생성 (중심 좌표는 데이터를 기준으로 설정)
    m = folium.Map(location=[filtered_data['Y좌표'].mean(), filtered_data['X좌표'].mean()], zoom_start=15)

    # CircleMarker를 사용하여 데이터 포인트 추가
    for idx, row in filtered_data.iterrows():
        folium.CircleMarker(
            location=[row['Y좌표'], row['X좌표']],
            radius=row[column_name] * 2,  # 원의 크기 조정
            color=row['color'],
            fill=True,
            fill_color=row['color'],
            fill_opacity=0.5,
            popup=f"노선번호: {row['노선번호']}<br>{time} {board_type}: {row[column_name]}"
        ).add_to(m)

    # Connect points in each group based on 정류장순서
    for route, group in filtered_data.groupby('노선번호'):
        points = group[['Y좌표', 'X좌표']].values.tolist()
        points.append(points[0])  # Connect last point back to the first point
        folium.PolyLine(
            locations=points,
            color=colors[route],
            weight=2.5,
            opacity=0.8,
            popup=f"Route: {route}"
        ).add_to(m)

    # 지도 저장
    map_file_name = f"kangseo_{time}_{board_type}_{day_type}.html"
    m.save(map_file_name)
    return m
```

## 승하차 총합 시각화

```python
def visualize_time_data(board_type, day_type):
    # 노선번호에 따라 색상 설정
    colors = {'강서01': 'red', '강서02': 'purple'}

    # 주어진 요일 필터링
    filtered_data = df[df['0평일_1토_2일(공)'] == day_type]

    # 노선번호에 따라 색상 설정
    filtered_data['color'] = filtered_data['노선번호'].map(colors)

    # 시간대별 평균 계산
    time_columns = [col for col in df.columns if board_type in col]
    filtered_data['average'] = filtered_data[time_columns].mean(axis=1)

    # Sort data within each group by 정류장순서
    filtered_data = filtered_data.sort_values(by=['노선번호', '정류장순서'])

    # folium 지도 생성 (중심 좌표는 데이터를 기준으로 설정)
    m = folium.Map(location=[filtered_data['Y좌표'].mean(), filtered_data['X좌표'].mean()], zoom_start=15)

    # CircleMarker를 사용하여 데이터 포인트 추가
    for idx, row in filtered_data.iterrows():
        folium.CircleMarker(
            location=[row['Y좌표'], row['X좌표']],
            radius=row['average'] * 2,  # 원의 크기 조정
            color=row['color'],
            fill=True,
            fill_color=row['color'],
            fill_opacity=0.5,
            popup=f"노선번호: {row['노선번호']}<br>평균 {board_type}: {row['average']:.2f}"
        ).add_to(m)

    # Connect points in each group based on 정류장순서
    for route, group in filtered_data.groupby('노선번호'):
        points = group[['Y좌표', 'X좌표']].values.tolist()
        points.append(points[0])  # Connect last point back to the first point
        folium.PolyLine(
            locations=points,
            color=colors[route],
            weight=2.5,
            opacity=0.8,
            popup=f"Route: {route}"
        ).add_to(m)

    # 지도 저장
    map_file_name = f"kangseo_average_{board_type}_{day_type}.html"
    m.save(map_file_name)
    return m
```
