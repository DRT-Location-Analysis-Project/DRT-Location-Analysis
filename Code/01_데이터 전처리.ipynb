```python
# 시간 컬럼 hh:mm 형식으로 변환
def convert_time(time):
    day, time_str = time.split('.')
    day = float(day)
    hours = float(time_str[:2])
    minutes = float(time_str[2:])
    total_hours = int(day * 24) + int(hours)
    minutes = int(minutes)
    return f"{total_hours:02d}:{minutes:02d}"
```

```python
# 시간 컬럼들
time_columns = ['첫차시간', '막차시간', '첫차시간(토요일)', '막차시간(토요일)', '첫차시간(공휴일)', '막차시간(공휴일)']

# 원하는 형식으로 변환
for col in time_columns:
    df_merge[col] = df_merge[col].apply(convert_time)

df_merge
```
